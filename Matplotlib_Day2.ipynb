# Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
data=np.random.randn(1000)
plt.hist(data,bins=20,edgecolor='red',color='pink',density=True,alpha=0.9)
plt.xlabel("data")
plt.grid()
plt.ylabel("values")
plt.title("Histogram Chart")
plt.show()

```

```python
import matplotlib.pyplot as plt
size=[12,25,26,27,89]
label=['a','b','c','d','e']
explode=[0,1,0,0,0]
color=['red','yellow','pink','blue','green','brown']  #'%.2%% for correct to 2 decimals
plt.axis('equal')
plt.pie(size,labels=label,colors=color,autopct='%.2f%%',startangle=80,shadow=True,explode=explode)
plt.title("pie chart")
```

# subplot

```python
import matplotlib.pyplot as plt
student=['anc', 'abc', 'pqr', 'xyz', '1mn']
maths=[85,56,69,52,46]
science=[95,75,85,45,62]

#subplot

fig,axs=plt.subplots (1,2, figsize=(12,5))

#plot maths
# the first axs is fo row and the second is for column
axs[0].bar(student, maths, color='b')

axs[0].set_title("Maths Score")

axs[0].set_xlabel("Student name")

axs[0].set_ylabel("maths")

#plot science

axs[1].bar(student, science,color='c')

axs[1].set_title("science Score")

axs[1].set_xlabel("Student name")

axs[1].set_ylabel("science")

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales = [4500, 5200, 6100, 4800, 5300, 5900, 6200, 7000, 6800, 7500, 8100, 8500]
monthly_profit = [900, 1200, 1500, 1300, 1100, 1400, 1700, 1900, 1800, 2100, 2400, 2600]

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # Assign subplots to axs

# Monthly sales subplot
axs[0].plot(months, monthly_sales, marker='o', color='b')
axs[0].set_title('Monthly Sales Trends')  
axs[0].set_xlabel('Months')  
axs[0].set_ylabel('Sales ')  
axs[0].grid(True)

# Monthly profit subplot
axs[1].plot(months, monthly_profit, marker='o', color='g')
axs[1].set_title('Monthly Profit Trends')  
axs[1].set_xlabel('Months')  
axs[1].set_ylabel('Profit ')  
axs[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
north_sales = [20000, 22000, 25000, 24000]
south_sales = [18000, 19000, 21000, 20000]
east_sales = [15000, 16000, 17000, 18000]
west_sales = [16000, 17000, 19000, 18000]
north_profit = [4000, 4500, 5000, 4700]
south_profit = [3500, 3800, 4200, 4000]
east_profit = [3000, 3200, 3500, 3300]
west_profit = [3200, 3500, 3700, 3400]

# Create subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# North region sales and profit
axs[0, 0].plot(quarters, north_sales, marker='o', color='b', label='Sales')
axs[0, 0].plot(quarters, north_profit, marker='o', color='g', label='Profit')
axs[0, 0].set_title('North Region')
axs[0, 0].set_xlabel('Quarters')
axs[0, 0].set_ylabel('Profit ')
axs[0, 0].legend()
axs[0, 0].grid(True)

# South region sales and profit
axs[0, 1].plot(quarters, south_sales, marker='o', color='b', label='Sales')
axs[0, 1].plot(quarters, south_profit, marker='o', color='g', label='Profit')
axs[0, 1].set_title('South Region')
axs[0, 1].set_xlabel('Quarters')
axs[0, 1].set_ylabel('Profit ')
axs[0, 1].legend()
axs[0, 1].grid(True)

# East region sales and profit
axs[1, 0].plot(quarters, east_sales, marker='o', color='b', label='Sales')
axs[1, 0].plot(quarters, east_profit, marker='o', color='g', label='Profit')
axs[1, 0].set_title('East Region')
axs[1, 0].set_xlabel('Quarters')
axs[1, 0].set_ylabel('Profit ')
axs[1, 0].legend()
axs[1, 0].grid(True)

# West region sales and profit
axs[1, 1].plot(quarters, west_sales, marker='o', color='b', label='Sales')
axs[1, 1].plot(quarters, west_profit, marker='o', color='g', label='Profit')
axs[1, 1].set_title('West Region')
axs[1, 1].set_xlabel('Quarters')
axs[1, 1].set_ylabel('Profit')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Quarterly performance scores for 8 quarters
sales_scores = [75, 82, 88, 90, 85, 78, 80, 83]
marketing_scores = [65, 70, 72, 68, 75, 77, 73, 69]
it_scores = [85, 88, 92, 89, 87, 91, 94, 90]
hr_scores = [60, 65, 63, 67, 70, 62, 66, 68]
finance_scores = [78, 80, 82, 79, 85, 87, 84, 81]

# Quarterly bonus amounts in thousands of dollars
sales_bonus = [5, 6, 7, 6, 6, 5, 6, 7]
marketing_bonus = [4, 5, 5, 4, 4, 5, 5, 4]
it_bonus = [7, 8, 9, 8, 8, 9, 9, 8]
hr_bonus = [3, 3.5, 3, 4, 4.5, 3.5, 3.5, 4]
finance_bonus = [6, 7, 7, 6, 7, 8, 7, 7]

departments = ['Sales', 'Marketing', 'IT', 'HR', 'Finance']
scores = [sales_scores, marketing_scores, it_scores, hr_scores, finance_scores]
bonuses = [sales_bonus, marketing_bonus, it_bonus, hr_bonus, finance_bonus]

# Create histograms for performance scores
plt.figure(figsize=(15, 10))

# Loop through each department by its index
for i in range(len(scores)):  # Using range to iterate over indices
    plt.subplot(2, 3, i+1)    # Create subplot at position i+1
    plt.hist(scores[i], bins=5, color='blue', edgecolor='black')  
    plt.title(departments[i] + ' Performance Scores')  
    plt.xlabel('Score')
    plt.ylabel('Frequency')

plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()

```
