# Simple bar chart __(barh -'h' is used for making it horizontaly)

```python
import matplotlib.pyplot as plt
x=[12,25,26,25,27]
y=['a','b','c','d','e']
plt.barh(x,y,color='blue')
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple bar plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data

categories= ['A', 'B', 'C', 'D']

# Data

values1 =[4, 7, 1, 8] 
values2 = [2, 5, 3, 6]

#Creating the bar chart

plt.bar(categories, values1, label='Set 1')

plt.bar(categories, values2, bottom=values1, label='Set 2') # Stacked on top of Set 1

#Adding labels and title

plt.xlabel('Categories')

plt.ylabel('Values')

plt.title('Stacked Bar Chart')

plt.legend()

#Display the chart

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
students = ['Ram', 'Aman', 'Raj', 'Avi']
math_scores = [36, 60, 64, 82]
science_scores = [68, 80, 60, 49]

# Number of students
num_students = len(students)

# Set up the positions for the groups
bar_width = 0.25
index = np.arange(num_students)


# Plot the bars for Math scores
plt.bar(index - bar_width/2, math_scores, bar_width, label='Math', color='blue')

# Plot the bars for Science scores
plt.bar(index + bar_width/2, science_scores, bar_width, label='Science', color='red')

# Add labels, title, and legend

plt.title('Scores of Students in Math and Science')

plt.xlabel('Students')

plt.ylabel('Scores')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

quarters = ['Q1', 'Q2', 'Q3', 'Q4']
values1 = [90, 180, 150, 200]
values2 = [100, 150, 90, 200]

# Plot the bars for Product X
plt.bar(quarters, values1, label='Product X', color='blue')

# Plot the bars for Product Y, stacked on top of Product X
plt.bar(quarters, values2, bottom=values1, label='Product Y', color='yellow')

# Add labels, title, and legend
plt.xlabel('Quarters')
plt.ylabel('Sales')
plt.title('Sales of Product X and Product Y Across Quarters')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [30, 24, 29, 23, 25, 20, 28]

# Create the line chart
plt.plot(days, temperatures, marker='o', color='b', linewidth=2)

# Add labels, title, and grid
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Vs Week')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue_a = [5000, 7000, 8000, 6000, 9000, 10000]
revenue_b = [4000, 6000, 7000, 5000, 8000, 9000]

# Create the line chart for Company A
plt.plot(months, revenue_a, marker='o', color='red', linestyle='-', linewidth=2, label='Company A')

# Create the line chart for Company B
plt.plot(months, revenue_b, marker='o', color='green', linestyle='--', linewidth=2, label='Company B')

# Add labels, title, and legend
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.title('Monthly Revenue of 6 months')
plt.legend()

# Show the plot
plt.show()
```
